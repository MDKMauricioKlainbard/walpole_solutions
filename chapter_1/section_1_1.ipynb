```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import trim_mean
# SOLVED EXERCISES FROM SECTION 1.1

# 1.1:
drying_hours = np.array([3.4, 2.5, 4.8, 2.9, 3.6, 2.8, 3.3, 5.6, 3.7, 2.8, 4.4, 4.0, 5.2, 3.0, 4.8])

# a)
sample_size = drying_hours.size
print(f"a) The sample size is: {sample_size}\n")

# b)
# I will start calculating things with loops, as if I were doing it by hand, to understand how it works.
# Afterwards, I will use NumPy methods exclusively to speed up the work.
suma = 0
for value in drying_hours:
    suma = suma + value
print(f"b) The sample mean is: {suma / sample_size}")
print(f"The sample mean calculated with NumPy is: {drying_hours.mean()}\n")

# c)
# I will take the median, depending on whether the sample size is even or odd.
median = 0
drying_hours_sorted = np.sort(drying_hours)
if sample_size % 2 == 0:
    index = np.int16((sample_size - 2) / 2)
    median = (drying_hours_sorted[index] + drying_hours_sorted[index+1]) / 2
else:
    index = np.int16((sample_size - 1) / 2)
    median = drying_hours_sorted[index]
print(f"c) The median is: {median}")
print(f"The median calculated with NumPy is: {np.median(drying_hours)}\n")

# d) I must use the Matplotlib library, so I’ll leave it for the end.

# e) I require the SciPy library
cut = np.int16(sample_size * 0.2)
trimmed_mean = trim_mean(drying_hours, proportiontocut=0.2)
trimmed_mean_manual = np.mean(drying_hours_sorted[cut:sample_size-cut])
print(f"e) The trimmed mean with SciPy is: {trimmed_mean}")
print(f"The trimmed mean, trimming the array manually and calculating with NumPy, is: {trimmed_mean_manual}")

# d)
y = [1] * len(drying_hours)  # all points will be on the same line

plt.figure(figsize=(8, 2))
plt.scatter(drying_hours, y, color="blue", s=100)

plt.yticks([])  # remove Y axis because it adds no information
plt.xlabel("Drying time (hours)")
plt.title("Dot plot - Drying times")
plt.show()

# f) The mean and the median are similar, but I can observe two outlier values somewhat far apart: 2.5 and 5.6.
# Therefore, in my personal judgment, I consider the median a better representative of this sample.
```

```python
# 1.2:
absorption_level = np.array([18.71, 21.41, 20.72, 21.81, 19.29, 22.43, 20.17, 23.71, 
                             19.44, 20.50, 18.92, 20.33, 23.00, 22.85, 19.25, 21.77, 
                             22.11, 19.77, 18.04, 21.12])

# a)
print(f"a) The sample mean is: {absorption_level.mean()}")
print(f"The sample median is: {np.median(absorption_level)}\n")

# b)
print(f"b) The 10% trimmed mean is: {trim_mean(absorption_level, proportiontocut=0.1)}")

# c)
plt.scatter(absorption_level, np.zeros(absorption_level.size), color="blue")

plt.yticks([])
plt.xlabel("Absorption level") 
plt.title("Dot plot - Absorption level")
plt.show()

# d) The values are very similar to each other, but both the mean and the trimmed mean
# tend to be slightly higher than the median. The extreme values are not very far
# from the center, but if we were to classify them as outliers, we could assume that
# there is one outlier on the right side of the center. In particular, 23.71.
```

```python
# 1.3:
# a)
without_aging = np.array([227, 222, 218, 217, 225, 218, 216, 229, 228, 221])
with_aging = np.array([219, 214, 215, 211, 209, 218, 203, 204, 201, 205])

y_with_aging = np.zeros(with_aging.size)
y_without_aging = y_with_aging + 1

plt.yticks([])
plt.scatter(with_aging, y_with_aging, color="blue", label="With aging")
plt.scatter(without_aging, y_without_aging, color="red", label="Without aging")
plt.xlabel("Tensile strength in PSI")
plt.title("Dot plot - Tensile strength in PSI of polymers")
plt.legend()
plt.show()

# b) From the dot plot, it can be inferred that the polymer bars without aging
# have higher tensile strength than those subjected to the accelerated
# aging process. In other words, the process had an effect.

# We justify this observation based on the fact that the data center of the
# non-aged set is greater than that of the aged set, in addition to a clear
# tendency for the data of the non-aged bars to be higher than the aged bars.

# c)
print(f"c) The mean of the aged bars is: {with_aging.mean()}")
print(f"The mean of the non-aged bars is: {without_aging.mean()}\n")

# d)
print(f"The median of the aged bars is: {np.median(with_aging)}")
print(f"The median of the non-aged bars is: {np.median(without_aging)}")

# The respective values of the means and medians are similar. We can assume that
# there are no outliers causing bias in the analysis. In fact, this was to be expected,
# considering that the experimental units are polymer bars, since there should not be
# too much variability between experimental units; it would be different if the unit
# were a human being, and it would be unethical to subject them to a tensile stress test.
```

```python
# 1.4:
company_A = np.array([9.3, 8.8, 6.8, 8.7, 8.5, 6.7, 8.0, 6.5, 9.2, 7.0])
company_B = np.array([11.0, 9.8, 9.9, 10.2, 10.1, 9.7, 11.0, 11.1, 10.2, 9.6])

y_A = np.zeros(company_A.size)
y_B = y_A + 1

plt.yticks([])
plt.scatter(company_A, y_A, c="red", label="Company A")
plt.scatter(company_B, y_B, c="blue", label="Company B")
plt.xlabel("Spring flexibility")
plt.title("Dot plot - Spring flexibility")
plt.legend()
plt.show()

# a)
print(f"The mean of Company A is: {company_A.mean()}, and its median is: {np.median(company_A)}")
print(f"The mean of Company B is: {company_B.mean()}, and its median is: {np.median(company_B)}")

# b) There is a more noticeable dispersion in the data from Company A, and overall,
# its springs have lower flexibility.
# The recorded data from Company B are less dispersed, and their values are generally higher.
# Possibly, the steel from Company A is of lower quality, as well as the process they use.
# Company B may have better quality control in this aspect.
```

```python
# 1.5:
control_group = np.array([7, 3, -4, 14, 2, 5, 22, -7, 9, 5])
treatment_group = np.array([-6, 5, 9, 4, 4, 12, 37, 5, 3, 3])

y_control = np.zeros_like(control_group)
y_treatment = y_control + 1
plt.yticks([])
plt.scatter(control_group, y_control, c="red", label="Control Group")
plt.scatter(treatment_group, y_treatment, c="green", label="Treatment Group")
plt.legend()
plt.show()

# b)
print("CONTROL GROUP")
print(f"Mean: {control_group.mean()}")
print(f"Median: {np.median(control_group)}")
print(f"Trimmed Mean 10%: {trim_mean(control_group, proportiontocut=0.1)}")
print("---")
print("TREATMENT GROUP")
print(f"Mean: {treatment_group.mean()}")
print(f"Median: {np.median(treatment_group)}")
print(f"Trimmed Mean 10%: {trim_mean(treatment_group, proportiontocut=0.1)}")

# c) The mean of the treatment group is higher than that of the control group,
# suggesting that the regimen might be effective. However, when analyzing the plot,
# we see the presence of an outlier, which is evidenced by the median and the trimmed mean.

# Disregarding this outlier, we observe a similar distribution between both groups,
# reflected in similar medians and trimmed means.
# This suggests that the treatment may not actually be effective.
```

```python
# 1.6:
lower_temperature = np.array([2.07, 2.14, 2.22, 2.03, 2.21, 2.03, 2.05, 2.18, 2.09, 2.14, 2.11, 2.02])
higher_temperature = np.array([2.52, 2.15, 2.49, 2.03, 2.37, 2.05, 1.99, 2.42, 2.08, 2.42, 2.29, 2.01])
y_l = np.zeros_like(lower_temperature)
y_h = y_l + 1

plt.yticks([])
plt.scatter(lower_temperature, y_l, c="blue", label="Temperature 20°C")
plt.scatter(higher_temperature, y_h, color="red", label="Temperature 45°C")
plt.xlabel("Tensile Strength")
plt.legend()
plt.title("Dot Plot - Tensile Strength")

# b)
print("TEMPERATURE 20°C")
print(f"Mean: {lower_temperature.mean()}")
print("---")
print("TEMPERATURE 45°C")
print(f"Mean: {higher_temperature.mean()}")

# c) Yes, it has an influence. However, it seems that high temperature causes some
# specimens to actually increase their tensile strength. Still, there seems to be
# a second factor involved.

# d) Speaking of this second factor related to temperature, it is possible that at a high
# temperature, although tensile strength can increase, there may be a certain limit
# that leads to material degradation, causing the opposite effect.
# Probably, at 45°C, some of these units manage to increase their strength,
# while others degrade significantly.
```

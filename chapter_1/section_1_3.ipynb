```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_sample_stadistics(np_array = [], title = ""):
    print("---\n")
    print(title)
    print(f"Mean: {np_array.mean()}")
    print(f"Median: {np.median(np_array)}")
    print(f"Variance: {np_array.var(ddof=1)}")
    print(f"Standard Deviation: {np_array.std(ddof=1)}")
    print(f"Range: {np_array.max() - np_array.min()}")
    print("\n---\n")
```

```python
# 1.13:

life_time = np.array([123, 116, 122, 110, 175, 126, 125, 111, 118, 117])
get_sample_stadistics(life_time, "LIFETIME OF THE COMPONENTS")

# There is a noticeable difference between the mean and the median (4.3). 
# In particular, we have a high standard deviation, which implies greater dispersion 
# of the data around the mean. This justifies the difference between both measures.

# In fact, it’s easy to point out the outlier: 175, which is considerably far from the mean.
```

```python
# 1.25
from scipy.stats import trim_mean

percentage_high_income_families = np.array([72.2, 31.9, 26.5, 29.1, 27.3, 8.6, 22.3, 26.5, 20.4, 12.8, 25.1, 19.2, 24.1, 58.2, 68.1, 89.2, 55.1, 9.4, 14.5, 13.9, 20.7, 17.9, 8.5, 55.4, 38.1, 54.2, 21.5, 26.2, 59.1, 43.3])

get_sample_stadistics(percentage_high_income_families, "PERCENTAGE HIGH INCOME FAMILIES")
print(f"10% trimmed mean: {trim_mean(percentage_high_income_families, proportiontocut=0.1)}")

# There is a significant difference between the mean and median in this case. 
# It is likely that schools with higher staff salaries per student have a higher proportion 
# of students from economically advantaged families. This might be expected, assuming 
# that higher salaries correspond to higher costs for attending the institution.

# We can assume that schools with lower salaries are "poorer" or less desirable. 
# Due to lower demand, students from families with fewer resources attend. 
# Consequently, the school’s budget is smaller, which is reflected in staff salaries.

# To verify, let’s create a scatter plot:
plt.title("Relationship between family income and staff salaries")
plt.scatter(percentage_high_income_families, income_dollars, label="Observations")
plt.xlabel("Percentage of families with higher income")
plt.ylabel("Staff income in dollars per student")
plt.legend()
plt.show()

# Observing the plot, the points tend to form a straight line. 
# It may be possible to define a functional relationship between the two variables, 
# of the type y = mx + b, where x is the percentage of high-income families and y is the salary.

# Stem-and-leaf plot
stemgraphic.stem_graphic(percentage_high_income_families)
```

```python
# 1.26

# Clearly, this type of analysis could have several weak points, as many factors are not taken into account.
# The salary analysis could be biased; for example, staff salaries may not be equitable. 
# Some staff members might earn significantly more than others.

# It could happen that a school has a high proportion of families with higher income, 
# but the budget is allocated to other purposes rather than maintaining high teacher salaries. 
# While atypical, it is possible.

# Another possibility is that a school is subsidized, allowing students from lower-income families to attend, 
# while salaries remain relatively high thanks to the subsidy.

# Note that we are ignoring that the functional relationship we are proposing is based on relating 
# the average staff salary in the education system to the average percentage of high-income families in the school system.

# Clearly, this introduces a significant bias. Assuming the typical family income level is not very high, 
# the mean would be skewed by prestigious schools attended by many high-income families. 
# In other words, the mean would suggest a higher proportion of high-income families in the school system, 
# resulting in an overestimated average teacher salary in the function, which would not be a good estimate.
```

```python
# 1.27:

average_wear = np.array([210, 325, 375])
load = np.array([700, 1000, 1300])

plt.title("Line Plot")
plt.plot(load, average_wear, label="Average bearing wear vs applied load")
plt.xlabel("Load")
plt.ylabel("Wear")
plt.legend()
plt.show()

# From the previous plot, one might infer a linear relationship between bearing wear and applied load. 
# However, looking at the individual sample data, there is also high variability. 
# If a true functional relationship exists, then plotting wear for each sample against the same load 
# should show the points clustering closely in a scatter plot.

wear_700 = np.array([145, 105, 260, 330])
wear_1000 = np.array([250, 195, 375, 480])
wear_1300 = np.array([150, 180, 420, 750])

plt.scatter(np.full_like(wear_700, fill_value=700), wear_700, c="red", label="Load 700 lb")
plt.scatter(np.full_like(wear_1000, fill_value=1000), wear_1000, c="green", label="Load 1000 lb")
plt.scatter(np.full_like(wear_1300, fill_value=1300), wear_1300, c="pink", label="Load 1300 lb")
plt.scatter(load, average_wear, label="Average wear vs load")
plt.legend()
plt.show()

# The points are widely dispersed around the mean. This dispersion is especially pronounced 
# at a load of 1300 lb, making the previously proposed functional relationship unclear. 
# The mean masks this variability.
# Now, what could be causing this notable dispersion? 
# Are the experimental units homogeneous? 
# Are the testing conditions consistent? 
# Is there an additional factor not being considered?
```

```python
# 1.31

low_speed_shrinkage = np.array([76.20, 76.09, 75.98, 76.15, 76.17, 75.94, 76.12, 76.18, 76.25, 75.82])
high_speed_shrinkage = np.array([93.25, 93.19, 92.87, 93.29, 93.37, 92.98, 93.47, 93.75, 93.89, 91.62])

y = np.zeros_like(low_speed_shrinkage)

plt.yticks([])
plt.scatter(low_speed_shrinkage, y, c="blue", label="Shrinkage at low injection speed")
plt.scatter(high_speed_shrinkage, y+1, c="red", label="Shrinkage at high injection speed")
plt.scatter(low_speed_shrinkage.mean(), 0, c="yellow", label="Mean shrinkage at low speed")
plt.scatter(high_speed_shrinkage.mean(), 1, c="pink", label="Mean shrinkage at high speed")
plt.xlabel("Shrinkage values (cm x 10^4)")
plt.title("Scatter plot")
plt.legend()
plt.show()

# If there were no interaction between temperature and injection speed, one would expect the same trend: 
# that high injection speed causes less shrinkage than low injection speed. However, here it is the opposite. 
# It becomes clear that high injection speed at high temperature causes much more shrinkage than low injection speed. 
# The trend completely changes, suggesting an interaction between injection speed and temperature.
```

```python
# 1.32

low_speed_shrinkage_low_temp = np.array([72.68, 72.62, 72.58, 72.48, 73.07, 72.55, 72.42, 72.84, 72.58, 72.92])
high_speed_shrinkage_low_temp = np.array([71.62, 71.68, 71.74, 71.48, 71.55, 71.52, 71.71, 71.56, 71.70, 71.50])

low_speed_shrinkage_high_temp = np.array([76.20, 76.09, 75.98, 76.15, 76.17, 75.94, 76.12, 76.18, 76.25, 75.82])
high_speed_shrinkage_high_temp = np.array([93.25, 93.19, 92.87, 93.29, 93.37, 92.98, 93.47, 93.75, 93.89, 91.62])

temperatures = np.array(["Low temperature", "High temperature"])

mean_low_speed = np.array([low_speed_shrinkage_low_temp.mean(), low_speed_shrinkage_high_temp.mean()])
mean_high_speed = np.array([high_speed_shrinkage_low_temp.mean(), high_speed_shrinkage_high_temp.mean()])

plt.plot(temperatures, mean_low_speed, c="blue", label="Shrinkage at low injection speed")
plt.plot(temperatures, mean_high_speed, c="red", label="Shrinkage at high injection speed")
plt.xlabel("Temperature")
plt.ylabel("Shrinkage (cm x 10^4)")
plt.legend()
plt.show()

# If there were no interaction between temperature and injection speed, the lines should be parallel. 
# Assuming a functional relationship of shrinkage with both factors without interaction, 
# a change in temperature at the same injection speed should cause a uniform shift of the line (additive effects).

# However, the plot of the means shows that the lines are not parallel, indicating that 
# there is indeed an interaction between the two factors.
```

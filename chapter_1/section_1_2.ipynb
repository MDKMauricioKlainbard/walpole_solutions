```python
import numpy as np
import matplotlib.pyplot as plt

# 1.7:
drying_hours = np.array([3.4, 2.5, 4.8, 2.9, 3.6, 2.8, 3.3, 5.6, 3.7, 2.8, 4.4, 4.0, 5.2, 3.0, 4.8])
# Once again, we will first perform the calculations using loops, then we will use libraries
# to speed up the results.
mean_value = drying_hours.mean()
sum = 0

for data in drying_hours:
    sum = sum + (mean_value - data)**2
    
variance = sum / (drying_hours.size - 1)
standard_deviation = variance ** 0.5

print("DRYING HOURS")
print(f"Mean: {mean_value}")
print(f"Variance: {variance}")
print(f"Variance with NumPy: {drying_hours.var(ddof=1)}")
print(f"Standard Deviation: {standard_deviation}")
print(f"Standard Deviation with NumPy: {drying_hours.std(ddof=1)}")

# We will plot the data again to visualize the results alongside the answers:
y = np.zeros_like(drying_hours)
plt.figure(figsize=(8, 2))
plt.scatter(drying_hours, y, color="blue")

plt.yticks([])  # remove Y axis because it adds no information
plt.xlabel("Drying time (hours)")
plt.title("Dot plot - Drying times")
plt.show()
```

```python
# 1.9:

without_aging = np.array([227, 222, 218, 217, 225, 218, 216, 229, 228, 221])
with_aging = np.array([219, 214, 215, 211, 209, 218, 203, 204, 201, 205])

y_with_aging = np.zeros(with_aging.size)
y_without_aging = y_with_aging + 1

plt.yticks([])
plt.scatter(with_aging, y_with_aging, color="blue", label="With aging")
plt.scatter(without_aging, y_without_aging, color="red", label="Without aging")
plt.xlabel("Tensile strength in PSI")
plt.title("Dot plot - Tensile strength in PSI of polymers")
plt.legend()
plt.show()

print("AGED BARS")
print(f"Mean: {with_aging.mean()}")
print(f"Variance: {with_aging.var()}")
print(f"Standard Deviation: {with_aging.std()}")
print("---")
print("NON-AGED BARS")
print(f"Mean: {without_aging.mean()}")
print(f"Variance: {without_aging.var()}")
print(f"Standard Deviation: {without_aging.std()}")

# There is an increase in the sample standard deviation of the aged bars.
# However, this increase is small compared to the magnitude of the values we are measuring.
# The tensile strengths are on the order of 10^2, while the change in standard deviation
# is on the order of 10^0.

# It is difficult to conclude that aging has an effect on the variability of the bars;
# for now, I leave that to my judgment.

# Assuming the experimental units were completely homogeneous, I would then conclude
# that yes, aging does have an effect on the variability of the tensile strengths.
```

```python
# 1.10:

company_A = np.array([9.3, 8.8, 6.8, 8.7, 8.5, 6.7, 8.0, 6.5, 9.2, 7.0])
company_B = np.array([11.0, 9.8, 9.9, 10.2, 10.1, 9.7, 11.0, 11.1, 10.2, 9.6])

y_A = np.zeros(company_A.size)
y_B = y_A + 1

plt.yticks([])
plt.scatter(company_A, y_A, c="red", label="Company A")
plt.scatter(company_B, y_B, c="blue", label="Company B")
plt.xlabel("Spring flexibility")
plt.title("Dot plot - Spring flexibility")
plt.legend()
plt.show()

def get_sample_stadistics(np_array = []):
    print(f"Mean: {np_array.mean()}")
    print(f"Variance: {np_array.var(ddof=1)}")
    print(f"Standard Deviation: {np_array.std(ddof=1)}")

print("SPRING FLEXIBILITY - COMPANY A")
get_sample_stadistics(company_A)
print("---")
print("SPRING FLEXIBILITY - COMPANY B")
get_sample_stadistics(company_B)

# As concluded before, Company A has a notable standard deviation.
# Additionally, its mean is lower than Company B's, while Company B maintains a smaller 
# standard deviation; the variability of its products is lower.
```

```python
# 1.11:

control_group = np.array([7, 3, -4, 14, 2, 5, 22, -7, 9, 5])
treatment_group = np.array([-6, 5, 9, 4, 4, 12, 37, 5, 3, 3])

y_control = np.zeros_like(control_group)
y_treatment = y_control + 1
plt.yticks([])
plt.scatter(control_group, y_control, c="red", label="Control Group")
plt.scatter(treatment_group, y_treatment, c="green", label="Treatment Group")
plt.legend()
plt.show()

print("CONTROL GROUP")
get_sample_stadistics(control_group)
print("---")
print("TREATMENT GROUP")
get_sample_stadistics(treatment_group)
```
